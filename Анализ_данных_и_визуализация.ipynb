# Гистограмма (Histogram)

Задача: Изучить распределение Цена за единицу.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
file_sales = "/content/drive/MyDrive/Анализ данных урок 3/df_sales.xlsx"  # Укажите путь к файлу
df_sales = pd.read_excel(file_sales, sheet_name="Sheet1")

# Построение гистограммы
plt.figure(figsize=(10, 6))
plt.hist(df_sales['Цена за единицу'], bins=10, edgecolor='black', color='skyblue')

# Настройки графика
plt.title('Распределение цены за единицу товара')
plt.xlabel('Цена за единицу')
plt.ylabel('Частота')
plt.grid(False)

# Отображение графика
plt.show()
```

**Вывод**

Основное количество товаров сосредоточено в диапазоне до 4000. При этом товаров с ценой выше 8000–10000 существует небольшое количество.
На стоимость 2000–3000 приходится большинство товаров. Это наиболее популярный ценовой сегмент. Более дешевые товары продаются в больших объемах.


Необходимо отметить, что товары с ценой выше 8000–10000 встречаются редко. Это может означать либо премиум-ассортимент, либо товары с низким спросом.

Выводы:

Большинство продаж приходится на товары с ценой до 4000, значит, на этот сегмент стоит делать упор в маркетинге и рекламе.

Можно проверить, какие товары находятся в дорогом сегменте (8000+) и как часто они продаются. Возможно, стоит пересмотреть их ассортимент или стратегию ценообразования.


# Ящик с усами (Boxplot)
Задача: Сравнить размах рейтинга товаров (Рейтинг товара) между разными категориями (Категория).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
file_sales = "/content/drive/MyDrive/Анализ данных урок 3/df_sales.xlsx"  # Укажите путь к файлу
df_sales = pd.read_excel(file_sales, sheet_name="Sheet1")

# Построение boxplot для сравнения размаха рейтинга товаров между категориями
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_sales, x="Категория", y="Рейтинг товара")

# Настройки графика
plt.xticks(rotation=45)
plt.title("Сравнение размаха рейтинга товаров по категориям")
plt.xlabel("Категория товара")
plt.ylabel("Рейтинг товара")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Отображение графика
plt.show()
```

Некоторые категории имеют широкий размах, что говорит о большой вариативности рейтингов товаров внутри этой категории.

**Футбол.** Медиана рейтинга находится достаточно высоко, что говорит о том, что большинство товаров в этой категории получают хорошие оценки. Размах небольшой, что говорит о стабильности оценок. Есть небольшое количество выбросов вниз, что может указывать на отдельные товары с низким качеством или недовольных клиентов.

**Бег.** Медиана рейтинга примерно на уровне средней оценки среди всех категорий. Наблюдается широкий размах – рейтинги варьируются от низких до высоких значений. Выбросы вниз показывают, что есть товары с очень низкими оценками. Возможно, проблема с качеством, комфортом или соответствием ожиданиям клиентов.

**Баскетбол.** Медиана чуть ниже, чем у Футбола, но все же ближе к высоким значениям.Разброс рейтингов не слишком большой, что говорит о том, что оценки товаров в этой категории более предсказуемы. Есть некоторые выбросы вниз, но их не так много, как в категории Бег.

Таким образом, анализ данных показывает, что Футбол – самая стабильная категория с высокими рейтингами. Бег – самая неоднородная категория. Стоит проверить товары с низкими оценками: возможно, они снижают средний рейтинг категории.
Баскетбол – относительно стабильная категория, но с некоторыми товарами, получающими низкие оценки. Можно провести анализ отзывов, чтобы понять причины.






# Столбчатая диаграмма (Bar Chart)
Задача: Показать суммарные продажи (общую сумму) по подкатегориям за весь период. Это даст представление о том, какие подкатегории наиболее доходны.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
file_sales = "/content/drive/MyDrive/Анализ данных урок 3/df_sales.xlsx"  # Укажите путь к файлу
df_sales = pd.read_excel(file_sales, sheet_name="Sheet1")

# Группировка данных по подкатегориям и сортировка по убыванию общей суммы продаж
subcategory_sales = df_sales.groupby('Подкатегория')['Общая сумма'].sum().reset_index()
subcategory_sales = subcategory_sales.sort_values(by='Общая сумма', ascending=False)

# Построение столбчатой диаграммы
plt.figure(figsize=(12, 8))
plt.bar(subcategory_sales['Подкатегория'], subcategory_sales['Общая сумма'], color='skyblue')

# Настройки графика
plt.title('Суммарные продажи по подкатегориям (отсортировано по убыванию)')
plt.xlabel('Подкатегория')
plt.ylabel('Общая сумма')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Отображение графика
plt.show()
```

Проанализировав этот график, можно сказать, что среди лидеров продаж - кроссовки. Эта категория значительно опережает остальные категории. Обувь также занимает сильную позицию, но уже с заметным отставанием.

На среднем уровне продаж мячи, форма и экипировка. Эти товары востребованы, но их продажи значительно ниже, чем у кроссовок.

Наименее доходные категории: шорты и футболки. Возможно, эти товары имеют более низкую стоимость или менее востребованы по сравнению с другими категориями.

Выводы:

Так как кроссовки и обувь приносят наибольший доход, стоит уделить им больше внимания в маркетинговых кампаниях, рекламе и акциях.


Можно изучить, почему форма и экипировка находятся на среднем уровне – возможно, увеличение ассортимента или изменения ценовой политики помогут повысить продажи.


Разобраться, почему шорты и футболки имеют низкие продажи. Это может быть связано с низким спросом, высокой конкуренцией или недостаточной рекламной поддержкой.

# Круговая диаграмма (Pie Chart)
Задача: Показать долю каждого способа оплаты (Способ оплаты) в общей сумме (Общая сумма).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
file_sales = "/content/drive/MyDrive/Анализ данных урок 3/df_sales.xlsx"  # Укажите путь к файлу
df_sales = pd.read_excel(file_sales, sheet_name="Sheet1")

# Группировка данных по способу оплаты и суммирование продаж
payment_distribution = df_sales.groupby("Способ оплаты")["Общая сумма"].sum().reset_index()

# Построение круговой диаграммы
plt.figure(figsize=(10, 6))
plt.pie(payment_distribution["Общая сумма"],
        labels=payment_distribution["Способ оплаты"],
        autopct='%1.1f%%',
        startangle=140)

# Настройки графика
plt.title("Распределение общей суммы продаж по способам оплаты")
plt.axis('equal')  # Делает круг идеальным

# Отображение графика
plt.show()
```

Анализ данных показывает, что карта – это самый популярный способ оплаты, на него приходится 37.3% общей суммы продаж.
Наличные занимают второе место с 32.2%. Онлайн-кошелек находится на третьем месте среди способов оплаты, но также имеет значительную долю 30.5%.


Необхолимо отметить, что продажи относительно равномерно распределены между способами оплаты, нет явного доминирования одного метода. Это говорит о том, что покупатели используют разные способы оплаты, и бизнесу важно обратить на это внимание и поддерживать все три метода.


В том случае, если карта и онлайн-платежи продолжают расти, стоит делать упор на удобство онлайн-оплаты (например, кэшбэк, бонусы).
Если наличные постепенно теряют популярность, можно предложить клиентам скидки за безналичный расчет.


# Линейная диаграмма (Line Chart)
Задача: Посмотреть, как изменяется среднее время доставки (Время доставки (дни)).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
file_sales = "/content/drive/MyDrive/Анализ данных урок 3/df_sales.xlsx"  # Укажите путь к файлу
df_sales = pd.read_excel(file_sales, sheet_name="Sheet1")

# Преобразование даты в формат datetime
df_sales["Дата продажи"] = pd.to_datetime(df_sales["Дата продажи"])

# Группировка данных по дате и вычисление среднего времени доставки
delivery_trend = df_sales.groupby("Дата продажи")["Время доставки (дни)"].mean().reset_index()

# Построение линейного графика
plt.figure(figsize=(12, 6))
sns.lineplot(data=delivery_trend, x="Дата продажи", y="Время доставки (дни)", marker="o", color="orange")

# Настройки графика
plt.title("Изменение среднего времени доставки по дням")
plt.xlabel("Дата продажи")
plt.ylabel("Среднее время доставки (дни)")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)

# Отображение графика
plt.show()
```

График показывает значительные колебания, время доставки варьируется от 1.5 дней (6 февраля) до 6 дней (19 января).
В некоторых датах наблюдаются резкие всплески, что может указывать на задержки или сбои в логистике.

Время доставки иногда снижается до 2.5–3 дней, что говорит о возможных периодах ускоренной обработки заказов.
В другие дни доставка достигает 6 дней, что может быть связано с внешними факторами (сезонные нагрузки, проблемы с поставками).

То есть, прослеживаются периоды, когда доставка ускоряется и замедляется. Возможно, задержки возникают в определённые дни недели или зависят от объемов заказов.

Выводы:
Необходимо проверить, в какие даты были максимальные задержки и что могло их вызвать (например, выходные, праздники, сбои в логистике). Если есть повторяющиеся пики, можно скорректировать распределение заказов или увеличить ресурсы в пиковые периоды.


# Воронка (Funnel)
Задача: Показать путь от «Посетили сайт» до «Получили товар». Используем  df_funnel. Построить воронку с помощью plotly

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
file_funnel = "/content/drive/MyDrive/Анализ данных урок 3/df_funnel.xlsx"  # Укажите путь к файлу
df_funnel = pd.read_excel(file_funnel, sheet_name="Sheet1")

# Очистка названий колонок от лишних пробелов и скрытых символов
df_funnel.columns = df_funnel.columns.str.strip()

# Инвертируем порядок шагов для корректного отображения (чтобы первый этап был сверху)
df_funnel = df_funnel[::-1]

# Построение воронки продаж
plt.figure(figsize=(8, 6))
plt.barh(df_funnel["Шаг"], df_funnel["Количество пользователей"], color="skyblue")

# Настройки графика
plt.xlabel("Количество пользователей")
plt.ylabel("Этап воронки")
plt.title("Воронка продаж")
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Отображение графика
plt.show()
```

Анализ воронки показывает, что большая потеря пользователей происходит на ранних этапах:

**Посетили сайт** 10 000 пользователей (это максимальное число посетителей).

**Зарегистрировались** 6 000 пользователей (то есть около 40% пользователей не регистрируются).

*Возможные причины*: неудобная регистрация, отсутствие мотивации зарегистрироваться, сложный интерфейс.

Далее мы наблюдаем снижение на этапе **Добавили в корзину** (4 000 пользователей): почти 30% зарегистрированных пользователей не добавляют товары в корзину.

*Возможные причины*: неинтересный ассортимент, неудобный поиск, недостаток информации о товарах.

Далее мы наблюдаем сокращение на этапе **Сформировали заказ** (2 500 пользователей): почти 40% пользователей, добавивших товары в корзину, не завершают заказ.

*Возможные причины*: высокая стоимость доставки, отсутствие скидок, сложный процесс оформления.

Следующий этап - **Оплатили** (около 1 500 пользователей):

Еще 1 000 пользователей не завершают оплату.

*Возможные причины:* нехватка удобных способов оплаты, технические ошибки, сомнения в надежности.

Последний этап – **Получили товар** (~1 000 пользователей):

Только около 10% от посетителей сайта доходят до покупки.
Однако от этапа оплаты до получения товара потери минимальны, что означает, что логистика работает стабильно.
